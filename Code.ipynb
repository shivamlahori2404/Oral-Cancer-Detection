```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model
import matplotlib.pyplot as plt

# Define directories for train, val, and test datasets
train_dir = '/kaggle/input/dataset/train'
val_dir = '/kaggle/input/dataset/val'
test_dir = '/kaggle/input/dataset/test'

# Preprocessing the train and validation data
train_datagen = ImageDataGenerator(rescale=1./255)
val_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

# Training data generator
train_data = train_datagen.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary'
)

# Validation data generator
validation_data = val_datagen.flow_from_directory(
    val_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary'
)

# Test data generator
test_data = test_datagen.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    shuffle=False  # Typically no shuffling in test set
)

# Load the pre-trained ResNet50 model for transfer learning
weights_path = '/kaggle/input/resnet/resnet50_weights_tf_dim_ordering_tf_kernels_notop.h5'  # Ensure this path is correct

# Load the base model with specified weights
base_model = ResNet50(weights=weights_path, include_top=False, input_shape=(224, 224, 3))

# Freezing the base model layers to retain pre-trained weights
for layer in base_model.layers:
    layer.trainable = False

# Adding custom layers on top of ResNet50
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(1024, activation='relu')(x)
x = Dense(512, activation='relu')(x)
predictions = Dense(1, activation='sigmoid')(x)

# Defining the final model
model = Model(inputs=base_model.input, outputs=predictions)

# Compiling the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Display the model summary
model.summary()

# Placeholder: Implement Aquila and Gorilla optimizers
class AquilaOptimizer:
    def __init__(self, params):
        self.params = params
    
    def optimize(self, objective_function):
        # Aquila optimization logic (Placeholder)
        return {"epochs": 50}  # Example parameter

class GorillaOptimizer:
    def __init__(self, params):
        self.params = params
    
    def optimize(self, objective_function):
        # Gorilla optimization logic (Placeholder)
        return {"epochs": 50}  # Example parameter

# Example objective function (optimizing model parameters)
def objective_function(params):
    # Adjust model parameters and return accuracy/loss
    return model.evaluate(validation_data)

# Initialize Aquila and Gorilla optimizers
params = {"learning_rate": 0.001, "epochs": 50}
aquila_opt = AquilaOptimizer(params)
gorilla_opt = GorillaOptimizer(params)

# Use optimizers to find the best parameters
best_params_aquila = aquila_opt.optimize(objective_function)
best_params_gorilla = gorilla_opt.optimize(objective_function)

# Model Training with the best params
history = model.fit(
    train_data,
    validation_data=validation_data,
    epochs=best_params_aquila["epochs"]
)

# Model Evaluation on Test Set
score = model.evaluate(test_data)
print('Test loss:', score[0])
print('Test accuracy:', score[1])

# Save the model
model.save('oral_cancer_detection_model.h5')

# Plot training & validation accuracy/loss
# Summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()

# Summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
```
